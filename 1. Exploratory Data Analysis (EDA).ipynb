# Conversion Rate Analysis Based on Income, Spending and Age

```python
# cek value web visit
df['NumWebVisitsMonth'].value_counts()
```

```python
# menghilangkan value 0 pada kolom web visit
df = df[df['NumWebVisitsMonth']>0]
```

## Conversion Rates by Age Group

```python
#mencari umur dengan cara operasi pengurangan Year_Birth dengan Dt_customer
df['years'] = df['Dt_Customer'].apply(lambda x : x.split('-')[2]) #split date time untuk tahun
df['age'] = np.int64(df['years']) - df['Year_Birth'] #np.int64 untuk convert tipe data string to integer

#melihat distirbusi umur
plt.figure(figsize=(15,5))
sns.distplot(df['age'])
sns.despine()
plt.title('Age Distribution')
plt.show()
```

```python
#membuat kategori berdasarkan umur
df['age_category'] = df['age'].apply(lambda x : 'Middle Ages' if x < 40 else ('Adults' if x < 60 else 'Elders'))

#table age to conversion rate
df_response = df[df['Response']==1].groupby('age_category').agg({'ID':'count'}).reset_index()
df_not_response = df[df['Response']==0].groupby('age_category').agg({'ID':'count'}).reset_index()
df_response['conversion_rate'] = df_response['ID'] / (df_response['ID'] + df_not_response['ID']) * 100
df_response['conversion_rate'] = df_response['conversion_rate'].round(3) #membulatkan angka dibekang koma menjadi 3 digits
dfg_age = df_response
dfg_age.rename(columns={'ID':'total_response'}, inplace=True)
dfg_age['age_category'] = pd.Categorical(dfg_age['age_category'], categories=['Middle Ages', 'Adults', 'Elders']) #mengurutkan value categories
dfg_age = dfg_age.sort_values('age_category', ascending=True)
dfg_age
```

```python
#plot
plt.figure(figsize=(10,5))
ax = sns.barplot(data=dfg_age, x=dfg_age['age_category'], y='conversion_rate', palette='pastel')
sns.despine()
plt.tight_layout()

#add labels
plt.bar_label(ax.containers[0], fmt='%.2f%%')

#add title
plt.title('Conversion Rate\nby Age Group', weight='bold', fontsize=15)
plt.xlabel('Age Group', fontsize=13)
plt.ylabel('Conversion Rate (%)', fontsize=13)
plt.show()
```

## Conversion Rates by Income Group

```python
# cek values incomes
df['Income'].describe().T
```

```python
# cek distribution incomes
plt.figure(figsize =(15,5))
sns.distplot(df['Income'], color='red')
sns.despine()
plt.tight_layout()

#add title
plt.title('Distribution of Income',weight='bold', fontsize=15)
plt.xlabel('Income (x100 millions)')

plt.xlim(0,667000000)
plt.show()
```

```python
# make category of incomes
df['incomes_category'] = df['Income'].apply(lambda x : 'low' if x < 51000000 else ('mid' if x < 68000000 else 'high'))
df['incomes_category'].value_counts()
```

```python
# table incomes to conversion rates
df_response = df[df['Response']==1].groupby(['incomes_category']).agg({'ID':'count'}).reset_index()
df_not_response = df[df['Response']==0].groupby(['incomes_category']).agg({'ID':'count'}).reset_index()
df_response['conversion_rate'] = df_response['ID'] / (df_response['ID'] + df_not_response['ID']) * 100
dfg_income = df_response
dfg_income['conversion_rate'] = dfg_income['conversion_rate'].round(3)
dfg_income.columns = ['incomes_category', 'count_response', 'conversion_rate'] #rename columns lebih dari 1 column
dfg_income['incomes_category'] = pd.Categorical(dfg_income['incomes_category'], categories=['low', 'mid', 'high'])
dfg_income = dfg_income.sort_values('incomes_category', ascending=True)
dfg_income
```

```python
#plot
plt.figure(figsize=(7,5))
ax = sns.barplot(data=dfg_income, x='incomes_category', y='conversion_rate', palette='pastel')
sns.despine()
plt.tight_layout()

#add barplot labels
plt.bar_label(ax.containers[0], fmt='%.2f%%')

#add title
plt.title('Conversion Rate\nby Income Group', weight='bold', fontsize=15)
plt.xlabel('Income Group', fontsize=13)
plt.ylabel('Conversion Rate (%)', fontsize=13)
plt.show()
```

## Conversion Rates by Spending Money

```python
# add total spending columns
df['total_spending'] = df['MntCoke'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']

# check values total spending
df['total_spending'].describe()
```

```python
# check distribution
plt.figure(figsize=(10,5))
sns.kdeplot(df['total_spending'], color='limegreen')
sns.despine()
plt.tight_layout()

#add title
plt.title('Distribution of Total Spending', weight='bold', fontsize=15)
plt.xlabel('total spending (x1 million)')
plt.xlim(0,2600000)
plt.show()
```

```python
# make category based on total spending
df['total_spending_category'] = df['total_spending'].apply(lambda x :'low' if x < 395000 else ('mid' if x < 1040000 else 'high'))
 
#cek value counts
df['total_spending_category'].value_counts()
```

```python
#make table group
df_response = df[df['Response']==1].groupby(df['total_spending_category']).agg({'ID':'count'}).reset_index()
df_not_response = df[df['Response']==0].groupby(df['total_spending_category']).agg({'ID':'count'}).reset_index()
df_response['conversion_rate'] = df_response['ID'] / (df_not_response['ID'] + df_response['ID'])*100
dfg_spending = df_response
dfg_spending.columns = ['total_spending_category','response','conversion_rate']
dfg_spending['total_spending_category'] = pd.Categorical(dfg_spending['total_spending_category'], categories = ['low', 'mid', 'high']) 
dfg_spending = dfg_spending.sort_values(by='total_spending_category', ascending=True)
dfg_spending['conversion_rate'] = dfg_spending['conversion_rate'].round(3)
```

```python
#plotting
plt.figure(figsize=(10,5))
ax = sns.barplot(data=dfg_spending, x='total_spending_category', y='conversion_rate', palette='pastel')
sns.despine()
plt.tight_layout()

#add bar label
plt.bar_label(ax.containers[0], fmt='%.2f%%')

#add title
plt.title('Conversion Rate by Spending Category')
plt.xlabel('Spending Category')
plt.ylabel('Conversion Rate (%)')
plt.show()
```
